```python
import math
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
```

```python

freq=100
min, max=-10, 10

xi=np.linspace(min, max, freq)
# y=[(-18*t*t*t+5*t*t + 10*t - 30)/(-12*t*t*t*t) for t in xi]
y=[-12*t*t*t*t*math.sin(math.cos(t)) - 18*t*t*t+5*t*t + 10*t - 30 for t in xi]
plt.plot(xi,y)
# plt.plot(x1,-35, 'ro')
plt.grid()
plt.show()
```
